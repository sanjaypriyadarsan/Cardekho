```python
import pandas as pd

bgl = pd.read_excel("C:/Users/User/Downloads/bangalore_cars.xlsx")
bgl['city'] = "Bangalore"

print(bgl)
chn = pd.read_excel("C:/Users/User/Downloads/chennai_cars.xlsx")
chn['city'] = "Chennai"
print(chn)
delhi = pd.read_excel("C:/Users/User/Downloads/delhi_cars.xlsx")
delhi['city']="Delhi"
print(delhi)
hyd = pd.read_excel("C:/Users/User/Downloads/hyderabad_cars.xlsx")
hyd['city'] ="Hyderabad"
print(hyd)
jaipur = pd.read_excel("C:/Users/User/Downloads/jaipur_cars.xlsx")
jaipur['city']="Jaipur"
print(jaipur)
kol = pd.read_excel("C:/Users/User/Downloads/kolkata_cars.xlsx")
kol['city']="kolkata"
print(kol)
df = pd.concat([bgl,chn,delhi,hyd,jaipur,kol], ignore_index =True)
print(df)
```

```python
import json
df.new_car_detail = df['new_car_detail'].apply(lambda x: json.loads(x.replace("'",'"').replace('None',"1")))
detail_df = pd.DataFrame(df['new_car_detail'].to_list())
detail_df
```

```python
import ast

# Convert stringified dictionaries to actual dictionaries
df.new_car_overview = df.new_car_overview.apply(ast.literal_eval)
def car_overview_dict_creater(overview):
  all_overview = {}
  for dictionary in overview['top']:
    all_overview.update({dictionary['key']:dictionary['value']})
  return all_overview
df.new_car_overview= df.new_car_overview.apply(lambda overview:car_overview_dict_creater(overview))
overview_df = pd.DataFrame(df.new_car_overview.to_list())
overview_df
```

```python
import ast

# Convert stringified dictionaries to actual dictionaries
df.new_car_specs = df.new_car_specs.apply(ast.literal_eval)
def car_specs_dict_creater(specifications):
  all_specifications = {}
  for dictionary in specifications['top']:
    all_specifications.update({dictionary['key']:dictionary['value']})
  return all_specifications
df.new_car_specs= df.new_car_specs.apply(lambda specifications:car_specs_dict_creater(specifications))
specifications_df = pd.DataFrame(df.new_car_specs.to_list())
specifications_df
```

```python
def convert_price(price_str):
    # Remove the currency symbol and whitespace
    price_str = str(price_str).replace('₹', '').replace(',', '').strip()
    
    # Check for Crore and convert accordingly
    if 'Crore' in price_str:
        value = price_str.replace(' Crore', '').strip()
        res = float(value)*10000000
    else:  # Assume it is in Lakh
        value = price_str.replace(' Lakh', '').strip()
        res = float(value)*100000
    return res

final_df['price_numeric'] = final_df['price'].apply(convert_price)

```

```python
import pandas as pd
import numpy as np

# Load dataset
# final_df = pd.read_csv("your_dataset.csv")  # Uncomment if loading from a CSV file

# Get numerical columns
num_cols = final_df.select_dtypes(include=['float64', 'int64']).columns

# Remove duplicate column names
num_cols = list(set(num_cols))

# Compute IQR and remove outliers
for col in num_cols:
    Q1 = final_df[col].quantile(0.25)
    Q3 = final_df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter outliers
    final_df = final_df[(final_df[col] >= lower_bound) & (final_df[col] <= upper_bound)]

# Display cleaned dataset
print(final_df.head())

# Save cleaned dataset if needed
# final_df.to_csv("cleaned_dataset.csv", index=False)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# Evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("****Linear Regression****")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R² Score: {r2}")
```

```python
# Initialize and train Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train.values.ravel())

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate performance
rf_mae = mean_absolute_error(y_test, y_pred_rf)
rf_mse = mean_squared_error(y_test, y_pred_rf)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, y_pred_rf)

print("****Random Forest Performance***")
print(f"MAE: {rf_mae}")
print(f"MSE: {rf_mse}")
print(f"RMSE: {rf_rmse}")
print(f"R² Score: {rf_r2}\n")
```

```python
# Initialize and train XGBoost model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
xgb_model.fit(X_train, y_train.values.ravel())

# Predictions
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate performance
xgb_mae = mean_absolute_error(y_test, y_pred_xgb)
xgb_mse = mean_squared_error(y_test, y_pred_xgb)
xgb_rmse = np.sqrt(xgb_mse)
xgb_r2 = r2_score(y_test, y_pred_xgb)

print("****XGBoost Performance****")
print(f"MAE: {xgb_mae}")
print(f"MSE: {xgb_mse}")
print(f"RMSE: {xgb_rmse}")
print(f"R² Score: {xgb_r2}\n")
```

```python
# Initialize and train Decision Tree model
dt_model = DecisionTreeRegressor(max_depth=10, random_state=42)
dt_model.fit(X_train, y_train)

# Predictions
y_pred_dt = dt_model.predict(X_test)

# Evaluate performance
dt_mae = mean_absolute_error(y_test, y_pred_dt)
dt_mse = mean_squared_error(y_test, y_pred_dt)
dt_rmse = np.sqrt(dt_mse)
dt_r2 = r2_score(y_test, y_pred_dt)

print("****Decision Tree Performance****")
print(f"MAE: {dt_mae}")
print(f"MSE: {dt_mse}")
print(f"RMSE: {dt_rmse}")
print(f"R² Score: {dt_r2}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define base model
rf = RandomForestRegressor(random_state=42)

# Define parameter grid
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform Grid Search
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid_rf, 
                              scoring='r2', cv=3, n_jobs=-1, verbose=2)
grid_search_rf.fit(X_train, y_train.values.ravel())

# Print best parameters
print("Best Parameters for Random Forest:", grid_search_rf.best_params_)

# Train with best parameters
best_rf = RandomForestRegressor(**grid_search_rf.best_params_, random_state=42)
best_rf.fit(X_train, y_train.values.ravel())

# Evaluate model
y_pred_rf = best_rf.predict(X_test)
print("Optimized Random Forest R² Score:", r2_score(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor

# Define base model
xgb = XGBRegressor(random_state=42)

# Define parameter grid
param_grid_xgb = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0]
}

# Perform Randomized Search (faster than GridSearch)
random_search_xgb = RandomizedSearchCV(estimator=xgb, param_distributions=param_grid_xgb, 
                                       scoring='r2', cv=3, n_iter=10, n_jobs=-1, verbose=2)
random_search_xgb.fit(X_train, y_train.values.ravel())

# Print best parameters
print("Best Parameters for XGBoost:", random_search_xgb.best_params_)

# Train with best parameters
best_xgb = XGBRegressor(**random_search_xgb.best_params_, random_state=42)
best_xgb.fit(X_train, y_train.values.ravel())

# Evaluate model
y_pred_xgb = best_xgb.predict(X_test)
print("Optimized XGBoost R² Score:", r2_score(y_test, y_pred_xgb))
```
